## üõ†Ô∏è PREPROCESAMIENTO DE DATOS

El preprocesamiento de datos es una fase fundamental en el desarrollo de modelos de machine learning. Esta etapa determina en gran medida la calidad y fiabilidad de nuestras predicciones, ya que transforma los datos crudos en un formato adecuado para el entrenamiento de modelos.

En esta secci√≥n desarrollaremos cuatro procesos esenciales:

* **An√°lisis estad√≠stico descriptivo**: Examinaremos las distribuciones, tendencias centrales y dispersi√≥n de nuestras variables
* **An√°lisis de balance de clases**: Evaluaremos la proporci√≥n entre las clases de nuestra variable objetivo
* **Estandarizaci√≥n**: Transformaremos nuestras variables num√©ricas para que est√©n en escalas comparables
* **Codificaci√≥n de variables categ√≥ricas**: Convertiremos las variables categ√≥ricas en formato num√©rico

Cada uno de estos pasos es cr√≠tico para garantizar que nuestros modelos puedan procesar los datos correctamente y generar predicciones confiables.

```python
# 1. Examinamos la informaci√≥n b√°sica del DataFrame
print("\n=== Informaci√≥n del DataFrame ===")
print(df.info())

# 2. Verificamos los tipos de datos y valores √∫nicos para cada columna
print("\n=== Valores √∫nicos por columna ===")
for column in df.columns:
    print(f"\n{column}:")
    print(f"Tipo de dato: {df[column].dtype}")
    print(f"Valores √∫nicos: {df[column].nunique()}")
    print(f"Primeros valores √∫nicos: {sorted(df[column].unique())[:5]}")

# 3. Verificamos valores nulos
print("\n=== Valores nulos ===")
print(df.isnull().sum())

# 4. Estad√≠sticas descriptivas b√°sicas
print("\n=== Estad√≠sticas descriptivas ===")
print(df.describe())

# 5. Debalance de clases

print("üìä Distribuci√≥n de la variable objetivo (default.payment.next.month):")
class_dist = df['default.payment.next.month'].value_counts(normalize=True)
print("\nPorcentajes:")
for clase, porcentaje in class_dist.items():
    print(f"Clase {clase}: {porcentaje*100:.2f}%")

# Visualizamos la distribuci√≥n con un gr√°fico de barras
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='default.payment.next.month')
plt.title('Distribuci√≥n de Defaults en la muestra')
plt.xlabel('Default (0: No, 1: S√≠)')
plt.ylabel('Cantidad')

# A√±adimos los valores exactos sobre cada barra
for i in plt.gca().containers[0]:
    plt.text(i.get_x() + i.get_width()/2,
            i.get_height(),
            f'{int(i.get_height())}',
            ha='center', va='bottom')

plt.show()
```

```python
# Funciones de Preprocesamiento de datos

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

def categorize_features(df):
    """
    Separa las caracter√≠sticas en num√©ricas y categ√≥ricas.
    """
    # Variables num√©ricas (montos y edad)
    numeric_features = [
        'LIMIT_BAL', 'AGE',
        'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
        'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'
    ]

    # Variables categ√≥ricas
    categorical_features = [
        'SEX', 'EDUCATION', 'MARRIAGE',
        'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6'
    ]

    return numeric_features, categorical_features

def clean_categorical_data(df):
    """
    Limpia y corrige valores en variables categ√≥ricas.
    """
    df = df.copy()

    # Corregir valores en EDUCATION
    # 0, 5 y 6 son valores no definidos, los convertimos a 4 (Others)
    df['EDUCATION'] = df['EDUCATION'].replace([0, 5, 6], 4)

    # Corregir valores en MARRIAGE
    # 0 es un valor no definido, lo convertimos a 3 (Others)
    df['MARRIAGE'] = df['MARRIAGE'].replace([0], 3)

    return df

def preprocess_data(df):
    """
    Realiza el preprocesamiento completo de los datos.
    """
    # 1. Separar features
    numeric_features, categorical_features = categorize_features(df)

    # 2. Limpiar datos categ√≥ricos
    df = clean_categorical_data(df)

    # 3. Estandarizar variables num√©ricas
    scaler = StandardScaler()
    df[numeric_features] = scaler.fit_transform(df[numeric_features])

    # 4. Codificar variables categ√≥ricas
    label_encoders = {}
    for feature in categorical_features:
        label_encoders[feature] = LabelEncoder()
        df[feature] = label_encoders[feature].fit_transform(df[feature].astype(str))

    # 5. Separar features y target
    X = df.drop('default.payment.next.month', axis=1)
    y = df['default.payment.next.month']

    # 6. Guardar informaci√≥n del preprocesamiento
    preprocessing_info = {
        'numeric_features': numeric_features,
        'categorical_features': categorical_features,
        'scaler': scaler,
        'label_encoders': label_encoders
    }

    return X, y, preprocessing_info

def print_preprocessing_summary(X, y):
    """
    Imprime un resumen del preprocesamiento.
    """
    print("\n=== Resumen del Preprocesamiento ===")
    print(f"Dimensiones de X: {X.shape}")

    print("\nDistribuci√≥n de clases:")
    print(pd.Series(y).value_counts(normalize=True).round(3))

    print("\nEstad√≠sticas de algunas variables num√©ricas:")
    print(X.describe().round(2).head())
```

```python
# Ejecutar el Preprocesamiento

# Importar y cargar datos
from datasets import load_dataset
import pandas as pd

# 1. Cargar datos
dataset = load_dataset("scikit-learn/credit-card-clients", streaming=True)
df = pd.DataFrame(list(dataset['train'].shuffle(seed=42).take(10000)))

# 2. Aplicar preprocesamiento
X, y, preprocessing_info = preprocess_data(df)

# 3. Imprimir resumen
print_preprocessing_summary(X, y)

# 4. Verificar las caracter√≠sticas procesadas
print("\n=== Caracter√≠sticas procesadas ===")
print("\nCaracter√≠sticas num√©ricas:")
print(preprocessing_info['numeric_features'])
print("\nCaracter√≠sticas categ√≥ricas:")
print(preprocessing_info['categorical_features'])

# 5. Mostrar informaci√≥n adicional sobre los datos procesados
print("\n=== Informaci√≥n adicional ===")
print(f"N√∫mero total de caracter√≠sticas: {X.shape[1]}")
print(f"N√∫mero total de observaciones: {X.shape[0]}")
print("\nPrimeras columnas del dataset procesado:")
print(X.head())
```

# üéØ ENTRENAMIENTO DE MODELOS DE CLASIFICACI√ìN

El entrenamiento de modelos es una fase cr√≠tica en el desarrollo de sistemas de machine learning para la predicci√≥n de riesgo crediticio. Esta etapa determina la capacidad de nuestros modelos para aprender patrones y realizar predicciones precisas sobre el comportamiento de pago de los clientes.

En esta secci√≥n desarrollaremos dos procesos fundamentales:**Preparaci√≥n para el entrenamiento** y **Proceso de entrenamiento**

Cada uno de estos pasos es esencial para desarrollar un sistema robusto de predicci√≥n de riesgo crediticio que pueda identificar efectivamente potenciales casos de default.

## Preparaci√≥n para el entrenamiento

Estableceremos el entorno controlado para nuestros experimentos mediante:
  - Fijaci√≥n de la semilla aleatoria para garantizar reproducibilidad
  - Divisi√≥n estratificada de datos en conjuntos de entrenamiento (80%) y prueba (20%)
  - Selecci√≥n de modelos de distancia.



```python
# --- PREPARACI√ìN PARA EL ENTRENAMIENTO ---

# Importamos las bibliotecas necesarias
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA
import numpy as np

# Fijamos la semilla aleatoria para garantizar reproducibilidad de resultados
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# Divisi√≥n estratificada de datos (80% entrenamiento, 20% prueba)
# stratify=y asegura que la proporci√≥n de las clases se mantenga en ambos conjuntos
X_train, X_test, y_train, y_test = train_test_split(
   X,              # Matriz de caracter√≠sticas
   y,              # Vector objetivo (default/no default)
   test_size=0.2,  # 20% para prueba
   random_state=RANDOM_STATE,  # Semilla para reproducibilidad
   stratify=y      # Mantiene proporci√≥n de clases
)

# Definici√≥n de los modelos supervisados a entrenar
models = {
    # K-Nearest Neighbors (KNN)
    # La elecci√≥n de k se basa en la ra√≠z cuadrada del n√∫mero de muestras
    "KNN": KNeighborsClassifier(
        n_neighbors=int(np.sqrt(len(X_train))),  # k basado en regla de la ra√≠z
        weights='uniform',       # Todos los vecinos tienen el mismo peso
        metric='euclidean'      # Distancia euclidiana para similitud
    ),

    # Support Vector Machine (SVM)
    # Utilizamos kernel RBF para capturar relaciones no lineales
    "SVM": SVC(
        kernel='rbf',           # Tipo de kernel: 'linear', 'poly', 'rbf', 'sigmoid'
        C=1.0,                  # Par√°metro de margen suave (soft margin):
                               # - C grande: margen m√°s estrecho, menos errores permitidos
                               # - C peque√±o: margen m√°s amplio, m√°s errores permitidos
        gamma='scale',          # Coeficiente Œ≥ para kernel RBF: exp(-Œ≥||x‚ÇÅ - x‚ÇÇ||¬≤)
                               # 'scale' calcula Œ≥ = 1 / (n_features * X.var())
        random_state=RANDOM_STATE
    ),

    # PCA + KNN
    # Combinamos reducci√≥n de dimensionalidad con clasificaci√≥n
    "PCA-KNN": {
        'pca': PCA(
            n_components=0.95,   # Mantener 95% de la varianza explicada
            random_state=RANDOM_STATE
        ),
        'knn': KNeighborsClassifier(
            n_neighbors=5,       # k inicial para clasificaci√≥n
            weights='uniform',
            metric='euclidean'
        )
    }
}

# Diccionarios para almacenar resultados del entrenamiento
predictions = {}      # Almacenar√° predicciones binarias (0/1)
probabilities = {}    # Almacenar√° probabilidades [0,1] donde sea aplicable
components = {}      # Almacenar√° componentes principales para PCA

# Imprimimos informaci√≥n sobre la divisi√≥n de datos
print("=== Informaci√≥n de la Divisi√≥n de Datos ===")
print(f"Dimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print("\nDistribuci√≥n de clases:")
print("\nConjunto de entrenamiento:")
print(pd.Series(y_train).value_counts(normalize=True).round(3))
print("\nConjunto de prueba:")
print(pd.Series(y_test).value_counts(normalize=True).round(3))
```

## Iniciar entrenamiento
  - Modelo KNN
  - Modelo Support Vector Machine
  - Modelo PCA-KNN

```python
# --- ENTRENAMIENTO DE MODELOS ---

def print_model_info(name, model, X_train):
    """
    Imprime informaci√≥n relevante seg√∫n el tipo de modelo
    """
    print(f"\nAn√°lisis de caracter√≠sticas para {name}:")

    if hasattr(model, 'coef_'):
        # Para modelos lineales que tienen coeficientes
        coef = pd.DataFrame({
            'Feature': X_train.columns,
            'Coefficient': model.coef_[0]
        }).sort_values('Coefficient', ascending=False)

        print("\nTop 5 caracter√≠sticas m√°s importantes (coeficientes m√°s altos):")
        print(coef.head())
        print("\nTop 5 caracter√≠sticas menos importantes (coeficientes m√°s bajos):")
        print(coef.tail())

    else:
        # Para modelos sin coeficientes
        if name == "KNN":
            print("KNN no tiene coeficientes al ser un algoritmo basado en distancias.")
            print(f"N√∫mero de vecinos utilizados: {model.n_neighbors}")
            print(f"M√©trica de distancia: {model.metric}")
        elif name == "SVM":
            print("Este modelo SVM no tiene coeficientes directamente interpretables.")
            print(f"Kernel utilizado: {model.kernel}")
            print(f"Par√°metro C: {model.C}")
        elif name == "K-means":
            print("K-means es un algoritmo de clustering no supervisado.")
            print(f"N√∫mero de clusters: {model.n_clusters}")
            print(f"M√©todo de inicializaci√≥n: {model.init}")
        elif name == "PCA-KNN":
            print("PCA-KNN es un pipeline de reducci√≥n de dimensionalidad + clasificaci√≥n.")
            print(f"Componentes PCA: {model['pca'].n_components}")
            print(f"N√∫mero de vecinos KNN: {model['knn'].n_neighbors}")

# Iteramos sobre cada modelo para entrenamiento y predicciones
for name, model in models.items():
    print(f"\n=== Entrenando modelo: {name} ===")

    # Entrenamiento del modelo
    if name == "K-means":
        # Para K-means, solo hacemos fit y guardamos las etiquetas de cluster
        model.fit(X_train)
        predictions[name] = model.labels_
        probabilities[name] = np.zeros(len(X_test))  # Placeholder
    elif name == "PCA-KNN":
        # Para PCA-KNN necesitamos crear un pipeline
        from sklearn.pipeline import Pipeline
        pipeline = Pipeline([
            ('pca', model['pca']),
            ('knn', model['knn'])
        ])
        pipeline.fit(X_train, y_train)
        predictions[name] = pipeline.predict(X_test)
        if hasattr(pipeline, 'predict_proba'):
            probabilities[name] = pipeline.predict_proba(X_test)[:, 1]
        else:
            probabilities[name] = predictions[name].astype(float)
        # Guardamos el pipeline en el diccionario de modelos para uso posterior
        models[name] = pipeline
    else:
        # Para modelos supervisados est√°ndar
        model.fit(X_train, y_train)
        predictions[name] = model.predict(X_test)
        if hasattr(model, 'predict_proba'):
            probabilities[name] = model.predict_proba(X_test)[:, 1]
        elif hasattr(model, 'decision_function'):
            probabilities[name] = model.decision_function(X_test)
        else:
            probabilities[name] = predictions[name].astype(float)

    # Mostrar informaci√≥n del modelo
    print_model_info(name, model, X_train)

# Calculamos y mostramos el tiempo de ejecuci√≥n promedio para predicciones
import time

print("\n=== Evaluaci√≥n de Tiempo de Predicci√≥n ===")
for name, model in models.items():
    times = []
    for _ in range(100):  # 100 predicciones para promedio estable
        start_time = time.time()
        if name == "K-means":
            model.predict(X_test[:100])
        else:
            model.predict(X_test[:100])
        times.append(time.time() - start_time)

    print(f"\n{name}:")
    print(f"Tiempo promedio de predicci√≥n: {np.mean(times)*1000:.2f} ms")
    print(f"Desviaci√≥n est√°ndar: {np.std(times)*1000:.2f} ms")
```

# üìä M√âTRICAS DE EVALUACI√ìN DE MODELOS DE RIESGO CREDITICIO

La evaluaci√≥n exhaustiva de modelos de clasificaci√≥n para riesgo crediticio requiere un an√°lisis desde m√∫ltiples perspectivas. Dividimos nuestras m√©tricas de evaluaci√≥n en dos categor√≠as principales: M√©tricas Tradicionales y Avanzadas.

## üéØ M√©tricas de Evaluaci√≥n Tradicionales

### Matriz de Confusi√≥n y sus Derivados (Para modelos supervisados)

#### 1. Accuracy (0 a 1)
- **Definici√≥n**: Proporci√≥n total de predicciones correctas
- **Valores objetivo**: > 0.7 suele ser aceptable
- **Consideraciones**:
  - Aunque √∫til, puede ser enga√±osa en datos desbalanceados como los nuestros
  - Ejemplo: Un accuracy de 0.95 indica que el 95% de todas las predicciones fueron correctas

#### 2. Precision (0 a 1)
- **Definici√≥n**: Del total de clientes que predecimos como morosos, cu√°ntos realmente lo son
- **Valores objetivo**: > 0.6 indican que cuando predecimos default, solemos estar en lo correcto
- **Importancia**:
  - Crucial para minimizar falsos positivos que podr√≠an llevar a rechazar buenos clientes
  - Ejemplo: Una precision de 0.8 significa que de cada 10 clientes que clasificamos como morosos, 8 realmente lo son

#### 3. Recall (0 a 1)
- **Definici√≥n**: Del total de clientes realmente morosos, cu√°ntos logramos identificar
- **Valores objetivo**: > 0.6 indican que detectamos bien los defaults reales
- **Importancia**:
  - Vital para minimizar p√©rdidas por defaults no detectados
  - Ejemplo: Un recall de 0.7 significa que detectamos el 70% de todos los defaults reales

#### 4. F1-Score (0 a 1)
- **Definici√≥n**: Media arm√≥nica entre Precision y Recall
- **Valores objetivo**: > 0.6 suele ser bueno
- **Caracter√≠sticas**:
  - Proporciona un balance entre ambas m√©tricas
  - Especialmente √∫til cuando buscamos un equilibrio entre detectar defaults y no rechazar buenos clientes

### Curva ROC y AUC (Para modelos supervisados)

#### Caracter√≠sticas principales:
- La curva ROC visualiza el balance entre verdaderos y falsos positivos a diferentes umbrales
- El AUC va de 0 a 1:
  - 0.5 indica un modelo aleatorio (l√≠nea diagonal)
  - > 0.7 se considera aceptable
  - > 0.8 se considera bueno
  - > 0.9 se considera excelente

#### 1. Silhouette Score (-1 a 1)
- **Definici√≥n**: Mide qu√© tan similar es un objeto a su propio cluster en comparaci√≥n con otros clusters
- **Interpretaci√≥n**:
  - -1: Agrupaci√≥n incorrecta
  - 0: Clusters solapados
  - 1: Clusters bien definidos y separados
- **Valores objetivo**: > 0.5 indica una buena separaci√≥n entre clusters

- Desarrollar estrategias espec√≠ficas para cada segmento
- Detectar comportamientos an√≥malos que podr√≠an indicar riesgo de default

```python
# --- Evaluaci√≥n y M√©tricas Tradicionales ---
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import (
    confusion_matrix, classification_report, accuracy_score,
    precision_score, recall_score, f1_score, roc_auc_score, roc_curve
)

class ModelEvaluator:
    def __init__(self, y_true, predictions, probabilities):
        """
        Inicializa el evaluador con los datos necesarios.

        Par√°metros:
        -----------
        y_true: array-like
            Etiquetas verdaderas
        predictions: dict
            Diccionario con predicciones de cada modelo
        probabilities: dict
            Diccionario con probabilidades de cada modelo
        """
        self.y_true = y_true

        # Asegurar consistencia en las dimensiones
        self.predictions = predictions
        self.probabilities = probabilities

        self.colors = ["#EDF2FB", "#ABC4FF"]

    def create_metrics_table(self):
        """
        Crea una tabla comparativa con las m√©tricas de todos los modelos.
        """
        metrics = []
        for name, pred in self.predictions.items():
            try:
                metric_dict = {
                    'Modelo': name,
                    'Accuracy': accuracy_score(self.y_true, pred),
                    'Precision': precision_score(self.y_true, pred, zero_division=0),
                    'Recall': recall_score(self.y_true, pred),
                    'F1-Score': f1_score(self.y_true, pred),
                    'AUC-ROC': roc_auc_score(self.y_true, self.probabilities[name])
                }
            except Exception as e:
                print(f"Error calculando m√©tricas para {name}: {str(e)}")
                metric_dict = {
                    'Modelo': name,
                    'Accuracy': 'Error',
                    'Precision': 'Error',
                    'Recall': 'Error',
                    'F1-Score': 'Error',
                    'AUC-ROC': 'Error'
                }

            metrics.append(metric_dict)

        df_metrics = pd.DataFrame(metrics)

        # Formatear las columnas num√©ricas
        numeric_columns = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC-ROC']

        for col in numeric_columns:
            if col in df_metrics.columns:
                df_metrics[col] = pd.to_numeric(df_metrics[col], errors='ignore')
                df_metrics[col] = df_metrics[col].apply(
                    lambda x: f"{x:.4f}" if isinstance(x, (int, float)) else x
                )

        return df_metrics

    def plot_confusion_matrices(self):
        """
        Visualiza las matrices de confusi√≥n para modelos de clasificaci√≥n.
        """
        n_models = len(self.predictions)
        fig, axes = plt.subplots(1, n_models, figsize=(5*n_models, 4))
        if n_models == 1:
            axes = [axes]

        cmap = sns.color_palette(self.colors, as_cmap=True)

        for idx, (name, y_pred) in enumerate(self.predictions.items()):
            try:
                cm = confusion_matrix(self.y_true, y_pred)
                sns.heatmap(
                    cm,
                    annot=True,
                    fmt='d',
                    ax=axes[idx],
                    cmap=cmap,
                    cbar=False,
                    linewidths=1,
                    linecolor='gray',
                    square=True
                )
                axes[idx].set_title(f'Matriz de Confusi√≥n\n{name}')
                axes[idx].set_xlabel('Predicci√≥n')
                axes[idx].set_ylabel('Valor Real')
                axes[idx].set_xticklabels(['No Default', 'Default'])
                axes[idx].set_yticklabels(['No Default', 'Default'], rotation=0)
            except Exception as e:
                print(f"Error al crear matriz de confusi√≥n para {name}: {str(e)}")

        plt.tight_layout()
        return fig

    def plot_roc_curves(self):
        """
        Visualiza las curvas ROC para modelos de clasificaci√≥n.
        """
        fig, ax = plt.subplots(figsize=(10, 6))
        colors = ['blue', 'orange', 'green', 'red']

        for (name, probs), color in zip(self.probabilities.items(), colors):
            try:
                fpr, tpr, _ = roc_curve(self.y_true, probs)
                auc = roc_auc_score(self.y_true, probs)
                ax.plot(fpr, tpr, color=color, lw=2,
                       label=f'{name} (AUC = {auc:.3f})')
            except Exception as e:
                print(f"Error al crear curva ROC para {name}: {str(e)}")

        ax.plot([0, 1], [0, 1], color='gray', linestyle='--')
        ax.set(xlim=[0.0, 1.0], ylim=[0.0, 1.05],
               xlabel='Tasa de Falsos Positivos',
               ylabel='Tasa de Verdaderos Positivos',
               title='Curvas ROC para los Modelos de Predicci√≥n de Default')
        ax.legend(loc="lower right")
        ax.grid(True, alpha=0.3)

        return fig

    def evaluate_all(self):
        """
        Ejecuta todas las evaluaciones y muestra los resultados.
        """
        # Tabla de m√©tricas
        try:
            metrics_table = self.create_metrics_table()
            print("\n=== Tabla Comparativa de M√©tricas ===")
            print(metrics_table.to_string(index=False))
        except Exception as e:
            print(f"Error al crear tabla de m√©tricas: {str(e)}")

        # Matrices de confusi√≥n
        print("\n=== Matrices de Confusi√≥n ===")
        cm_fig = self.plot_confusion_matrices()
        if cm_fig:
            plt.show()

        # Curvas ROC
        print("\n=== Curvas ROC ===")
        roc_fig = self.plot_roc_curves()
        if roc_fig:
            plt.show()

# Uso del evaluador
evaluator = ModelEvaluator(y_test, predictions, probabilities)
evaluator.evaluate_all()
```

### üéØ Entrop√≠a
La entrop√≠a en machine learning mide la incertidumbre o "duda" en las predicciones de nuestros modelos. Un valor de entrop√≠a:
- Cercano a 0: El modelo est√° muy seguro de sus predicciones
- Cercano a 1: El modelo tiene mucha incertidumbre
- El m√°ximo para clasificaci√≥n binaria es ln(2) ‚âà 0.693



```python
# --- An√°lisis de Entrop√≠a para Modelos de Clasificaci√≥n ---

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import entropy

def calculate_entropy(probabilities):
    """
    Calcula la entrop√≠a de las predicciones para clasificaci√≥n binaria.

    Par√°metros:
    -----------
    probabilities: array-like
        Array de probabilidades de la clase positiva (default)

    Retorna:
    --------
    array-like
        Valores de entrop√≠a para cada predicci√≥n
    """
    # Convertir a probabilidades de ambas clases
    p = np.vstack([1 - probabilities, probabilities]).T
    return entropy(p, axis=1)

def analyze_entropy(y_true, probabilities):
    """
    Analiza y visualiza la entrop√≠a de las predicciones de los modelos de clasificaci√≥n.

    Par√°metros:
    -----------
    y_true: array-like
        Etiquetas verdaderas
    probabilities: dict
        Diccionario con las probabilidades predichas por cada modelo
    """
    # Calcular entrop√≠a para cada modelo
    entropy_dict = {}
    mean_entropy_dict = {}

    # Verificar que hay modelos para analizar
    if not probabilities:
        print("No hay modelos para analizar la entrop√≠a.")
        return

    for name, probs in probabilities.items():
        try:
            entropy_vals = calculate_entropy(probs)
            entropy_dict[name] = entropy_vals
            mean_entropy_dict[name] = np.mean(entropy_vals)
            print(f"\nEntrop√≠a media para {name}: {mean_entropy_dict[name]:.4f}")
        except Exception as e:
            print(f"Error calculando entrop√≠a para {name}: {str(e)}")
            continue

    # Crear subplot con dos gr√°ficas
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # 1. Distribuci√≥n de entrop√≠a
    colors = ['blue', 'orange', 'green']
    for (name, entropy_vals), color in zip(entropy_dict.items(), colors):
        sns.kdeplot(data=entropy_vals, label=name, ax=ax1, color=color)
    ax1.set_title('Distribuci√≥n de Entrop√≠a por Modelo')
    ax1.set_xlabel('Entrop√≠a')
    ax1.set_ylabel('Densidad')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # 2. Entrop√≠a vs Probabilidad predicha
    for (name, entropy_vals), color in zip(entropy_dict.items(), colors):
        ax2.scatter(probabilities[name], entropy_vals,
                   alpha=0.5, label=name, color=color)
    ax2.set_title('Entrop√≠a vs Probabilidad de Default')
    ax2.set_xlabel('Probabilidad de Default')
    ax2.set_ylabel('Entrop√≠a')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # An√°lisis adicional: Entrop√≠a por clase real
    print("\nEntrop√≠a media por clase real:")
    for name, entropy_vals in entropy_dict.items():
        print(f"\n{name}:")
        for clase in [0, 1]:
            mask = y_true == clase
            mean_entropy = np.mean(entropy_vals[mask])
            print(f"Clase {clase} ({'No Default' if clase == 0 else 'Default'}): {mean_entropy:.4f}")

# Ejecutar el an√°lisis
print("=== An√°lisis de Entrop√≠a para Modelos de Clasificaci√≥n ===")
analyze_entropy(y_test, probabilities)
```

## üéØ Validaci√≥n Cruzada en Riesgo Crediticio: Un Enfoque Robusto


La validaci√≥n cruzada es una t√©cnica estad√≠stica que nos permite evaluar la capacidad real de nuestros modelos para predecir nuevos casos. En lugar de confiar en una √∫nica divisi√≥n de datos de entrenamiento y prueba, la validaci√≥n cruzada divide sistem√°ticamente nuestros datos en m√∫ltiples subconjuntos, entrena y eval√∫a el modelo varias veces, proporcionando una estimaci√≥n m√°s robusta y confiable de su rendimiento.


## üîÑ Enfoques de Validaci√≥n Cruzada que Implementaremos

### 1. Validaci√≥n Cruzada K-Fold üìä
- Divide los datos en K partes iguales
- Entrena el modelo K veces, usando cada parte como conjunto de prueba una vez
- Nos da una visi√≥n general de la estabilidad del modelo
- Revela qu√© tan bien generaliza nuestro modelo a diferentes subconjuntos de datos

### 2. Validaci√≥n Cruzada Estratificada üìà
- Similar al K-Fold, pero mantiene la proporci√≥n de clases en cada divisi√≥n
- Especialmente importante dado nuestro desbalance entre buenos pagadores y defaults
- Proporciona una evaluaci√≥n m√°s realista del rendimiento en casos de default
- Ayuda a entender mejor c√≥mo maneja el modelo los casos minoritarios pero cr√≠ticos

## üìä Curvas de Aprendizaje: La Historia de C√≥mo Aprende Nuestro Modelo

Las curvas de aprendizaje son como el diario de desarrollo de nuestro modelo, cont√°ndonos la historia de c√≥mo mejora (o no) a medida que le proporcionamos m√°s datos. En el contexto de riesgo crediticio, estas curvas son especialmente reveladoras, ya que nos ayudan a entender si nuestro modelo est√° realmente aprendiendo a identificar patrones de riesgo de manera efectiva.




```python
# Importaci√≥n de bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, StratifiedKFold, cross_validate, learning_curve
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score

# Configuraci√≥n global del ambiente
# RANDOM_STATE asegura reproducibilidad en los resultados
RANDOM_STATE = 42

# Configuraci√≥n del estilo de visualizaci√≥n de matplotlib
plt.rcParams.update({
    'figure.autolayout': True,  # Ajuste autom√°tico del layout
    'axes.grid': True,          # Mostrar cuadr√≠cula en los gr√°ficos
    'grid.alpha': 0.3           # Transparencia de la cuadr√≠cula
})

def get_cross_validation_results(models, X, y, k=5):
    """
    Realiza validaci√≥n cruzada K-Fold y Estratificada para m√∫ltiples modelos.

    Par√°metros:
    -----------
    models : dict
        Diccionario con los modelos a evaluar {nombre_modelo: modelo}
    X : array-like
        Caracter√≠sticas del dataset
    y : array-like
        Variable objetivo
    k : int, default=5
        N√∫mero de folds para la validaci√≥n cruzada

    Retorna:
    --------
    tuple
        (resultados_kfold, resultados_estratificados)
        Cada elemento contiene los scores de las m√©tricas para cada modelo
    """
    # Definici√≥n de m√©tricas de evaluaci√≥n
    scoring = {
        'accuracy': 'accuracy',                          # Exactitud general
        'precision': make_scorer(precision_score),       # Precisi√≥n (verdaderos positivos)
        'recall': make_scorer(recall_score),            # Recall (sensibilidad)
        'f1': make_scorer(f1_score)                     # F1-Score (media arm√≥nica)
    }

    # Inicializaci√≥n de diccionarios para almacenar resultados
    kfold_results = {}
    stratified_results = {}

    # Creaci√≥n de objetos de validaci√≥n cruzada
    kfold = KFold(n_splits=k, shuffle=True, random_state=RANDOM_STATE)
    stratified = StratifiedKFold(n_splits=k, shuffle=True, random_state=RANDOM_STATE)

    # Iteraci√≥n sobre cada modelo para realizar validaci√≥n cruzada
    for name, model in models.items():
        # Validaci√≥n K-Fold est√°ndar
        kfold_results[name] = cross_validate(
            model, X, y,
            scoring=scoring,
            cv=kfold,
            return_train_score=True  # Tambi√©n obtener scores de entrenamiento
        )

        # Validaci√≥n estratificada (mantiene proporci√≥n de clases)
        stratified_results[name] = cross_validate(
            model, X, y,
            scoring=scoring,
            cv=stratified,
            return_train_score=True
        )

    return kfold_results, stratified_results

def plot_boxplots_comparison(kfold_results, stratified_results):
    """
    Genera visualizaci√≥n comparativa de boxplots para ambos tipos de validaci√≥n cruzada.

    Par√°metros:
    -----------
    kfold_results : dict
        Resultados de la validaci√≥n cruzada K-Fold
    stratified_results : dict
        Resultados de la validaci√≥n cruzada estratificada

    Genera:
    -------
    Figura con 8 subplots (4 m√©tricas x 2 tipos de CV)
    """
    # Definici√≥n de m√©tricas a visualizar
    metrics = ['test_accuracy', 'test_precision', 'test_recall', 'test_f1']

    # Creaci√≥n de la figura principal
    fig = plt.figure(figsize=(20, 16))
    plt.subplots_adjust(top=0.9)  # Ajuste para el t√≠tulo principal

    # Generaci√≥n de subplots para cada m√©trica y tipo de CV
    for idx, metric in enumerate(metrics):
        # Subplot para K-Fold
        ax1 = plt.subplot(4, 2, 2*idx + 1)
        data_kfold = [kfold_results[model][metric] for model in kfold_results.keys()]
        ax1.boxplot(data_kfold)
        ax1.set_xticklabels(list(kfold_results.keys()), rotation=45)
        ax1.set_title(f'K-Fold CV: {metric.replace("test_", "").capitalize()}')
        ax1.grid(True, alpha=0.3)

        # Subplot para CV Estratificada
        ax2 = plt.subplot(4, 2, 2*idx + 2)
        data_strat = [stratified_results[model][metric] for model in stratified_results.keys()]
        ax2.boxplot(data_strat)
        ax2.set_xticklabels(list(stratified_results.keys()), rotation=45)
        ax2.set_title(f'Stratified CV: {metric.replace("test_", "").capitalize()}')
        ax2.grid(True, alpha=0.3)

    # Ajustes finales de la figura
    plt.suptitle('Comparaci√≥n de M√©todos de Validaci√≥n Cruzada', y=0.98, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

def plot_learning_curves_comparison(models, X, y):
    """
    Genera curvas de aprendizaje comparativas para m√∫ltiples modelos.

    Par√°metros:
    -----------
    models : dict
        Diccionario con los modelos a evaluar {nombre_modelo: modelo}
    X : array-like
        Caracter√≠sticas del dataset
    y : array-like
        Variable objetivo

    Genera:
    -------
    Figura con tres subplots, uno para cada modelo
    """
    # Configuraci√≥n de la validaci√≥n cruzada estratificada
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

    # Creaci√≥n de la figura
    fig = plt.figure(figsize=(20, 6))

    # Generaci√≥n de curvas de aprendizaje para cada modelo
    for idx, (name, model) in enumerate(models.items(), 1):
        plt.subplot(1, 3, idx)

        # C√°lculo de las curvas de aprendizaje
        train_sizes, train_scores, val_scores = learning_curve(
            model, X, y,
            train_sizes=np.linspace(0.1, 1.0, 10),  # 10 puntos entre 10% y 100% del dataset
            cv=cv,
            scoring='f1',    # Usando F1-score como m√©trica
            n_jobs=-1        # Usar todos los n√∫cleos disponibles
        )

        # C√°lculo de estad√≠sticas
        train_mean = np.mean(train_scores, axis=1)
        train_std = np.std(train_scores, axis=1)
        val_mean = np.mean(val_scores, axis=1)
        val_std = np.std(val_scores, axis=1)

        # Visualizaci√≥n de las curvas
        plt.plot(train_sizes, train_mean, label='Training score', color='blue')
        plt.plot(train_sizes, val_mean, label='Cross-validation score', color='red')

        # A√±adir bandas de desviaci√≥n est√°ndar
        plt.fill_between(train_sizes, train_mean - train_std,
                        train_mean + train_std, alpha=0.1, color='blue')
        plt.fill_between(train_sizes, val_mean - val_std,
                        val_mean + val_std, alpha=0.1, color='red')

        # Configuraci√≥n del subplot
        plt.title(f'Learning Curves\n{name}')
        plt.xlabel('Training Examples')
        plt.ylabel('F1 Score')
        plt.legend(loc='lower right')
        plt.grid(True, alpha=0.3)

    # Ajustes finales de la figura
    plt.suptitle('Curvas de Aprendizaje por Modelo', y=1.05, fontsize=16)
    plt.tight_layout()
    plt.show()

```

### Boxplot de las t√©cnicas de CV

# üéØ K-means Clustering: An√°lisis No Supervisado de Textos Educativos

En esta pr√°ctica, exploraremos t√©cnicas de clustering para analizar y organizar autom√°ticamente contenido educativo. Utilizaremos el dataset [PersonaHub FineWeb-Edu](https://huggingface.co/datasets/argilla/FinePersonas-v0.1-clustering-100k) de Hugging Face, que contiene una amplia colecci√≥n de textos educativos. Nuestro objetivo es descubrir patrones naturales en estos datos mediante dos enfoques de clustering: K-means est√°ndar y una combinaci√≥n de PCA con K-means. A trav√©s de este an√°lisis, aprenderemos c√≥mo las t√©cnicas de aprendizaje no supervisado pueden ayudar a organizar y estructurar grandes vol√∫menes de contenido educativo.

## üìö Sobre el Dataset y Contexto

El dataset PersonaHub FineWeb-Edu representa una colecci√≥n estructurada de 100,000 ejemplos de contenido educativo, procesado mediante una sofisticada pipeline de an√°lisis de texto. Este conjunto de datos fue creado utilizando distilabel 1.4.0 y ha pasado por varias etapas de procesamiento:

1. **Vectorizaci√≥n**: Transformaci√≥n de textos en embeddings de 1024 dimensiones
2. **Reducci√≥n**: UMAP reduce estos vectores a 2 dimensiones
3. **Agrupaci√≥n**: DBSCAN identifica 177 clases distintas
4. **Etiquetado**: Cada cluster recibe tres etiquetas sem√°nticas

Esta rica estructura de datos nos permite aplicar t√©cnicas de clustering para descubrir patrones y relaciones significativas en el contenido educativo.

## üîÑ Metodolog√≠a Implementada

Nuestro an√°lisis implementa dos enfoques complementarios:

### 1. K-means Est√°ndar
- Aplicado directamente sobre las variables num√©ricas estandarizadas
- Utiliza distancia euclidiana para agrupar observaciones similares
- Permite visualizaci√≥n mediante t-SNE para datos de alta dimensionalidad

### 2. PCA + K-means
- Reduce la dimensionalidad antes de aplicar clustering
- Mantiene la informaci√≥n m√°s relevante de los datos
- Facilita la visualizaci√≥n directa de clusters en 2D
- Especialmente √∫til cuando trabajamos con representaciones vectoriales de texto

## üìä M√©tricas de Evaluaci√≥n

Para evaluar la calidad del clustering, utilizamos:

1. **M√©todo del Codo**:
   - Ayuda a determinar el n√∫mero √≥ptimo de clusters
   - Muestra el punto donde agregar m√°s clusters no aporta beneficio significativo
   - En datos textuales, esto puede indicar el n√∫mero natural de categor√≠as tem√°ticas

2. **Silhouette Score**:
   - Mide qu√© tan similar es un objeto a su propio cluster vs. otros clusters
   - Rango: [-1, 1], donde valores m√°s altos indican mejor clustering
   - Particularmente √∫til para evaluar la coherencia de los grupos textuales


```python
# --- AN√ÅLISIS DE CLUSTERING CON K-MEANS ---

# ===== IMPORTACI√ìN DE BIBLIOTECAS =====
# Manipulaci√≥n de datos
import pandas as pd
import numpy as np
from datasets import load_dataset

# Preprocesamiento y modelos
from sklearn.preprocessing import StandardScaler  # Para estandarizar variables
from sklearn.decomposition import PCA            # Para reducci√≥n de dimensionalidad
from sklearn.cluster import KMeans               # Para clustering
from sklearn.metrics import silhouette_score     # Para evaluar calidad del clustering

# Visualizaci√≥n
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE                # Para visualizaci√≥n de alta dimensionalidad

# Configuraci√≥n inicial
plt.style.use('default')  # Estilo b√°sico de matplotlib
RANDOM_STATE = 42         # Semilla para reproducibilidad

def load_and_preprocess_data():
    """
    Carga y preprocesa los datos del dataset FinePersonas.

    Returns:
        tuple: (X_scaled, feature_names)
            - X_scaled: datos estandarizados
            - feature_names: nombres de las caracter√≠sticas
    """
    # 1. Carga de datos usando streaming de Hugging Face
    print("Cargando datos...")
    dataset = load_dataset("scikit-learn/credit-card-clients", streaming=True)
    df = pd.DataFrame(list(dataset['train'].shuffle(seed=42).take(10000)))

    # 2. Selecci√≥n de caracter√≠sticas num√©ricas
    numeric_features = df.select_dtypes(include=['float64', 'int64']).columns
    X = df[numeric_features]

    # 3. Estandarizaci√≥n (media=0, desviaci√≥n=1)
    print("Estandarizando variables...")
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Imprimir informaci√≥n sobre las dimensiones de los datos
    print(f"\nInformaci√≥n del dataset:")
    print(f"N√∫mero de muestras: {X_scaled.shape[0]}")
    print(f"N√∫mero de caracter√≠sticas: {X_scaled.shape[1]}")
    print(f"Caracter√≠sticas disponibles: {', '.join(X.columns)}\n")

    return X_scaled, X.columns

def find_optimal_clusters(X, max_clusters=10):
    """
    Encuentra el n√∫mero √≥ptimo de clusters usando dos m√©todos:
    1. M√©todo del codo
    2. Silhouette score

    Args:
        X: matriz de datos estandarizados
        max_clusters: n√∫mero m√°ximo de clusters a probar

    Returns:
        tuple: (inertias, silhouette_scores)
    """
    inertias = []          # Para m√©todo del codo
    silhouette_scores = [] # Para score de silueta

    print("Evaluando diferentes n√∫meros de clusters...")
    for k in range(2, max_clusters + 1):
        # Entrenamiento de K-means
        kmeans = KMeans(n_clusters=k, random_state=RANDOM_STATE)
        kmeans.fit(X)

        # Guardar m√©tricas
        inertias.append(kmeans.inertia_)
        silhouette_scores.append(silhouette_score(X, kmeans.labels_))

    # Visualizaci√≥n de resultados
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Gr√°fico del m√©todo del codo
    ax1.plot(range(2, max_clusters + 1), inertias, marker='o')
    ax1.set_title('M√©todo del Codo')
    ax1.set_xlabel('N√∫mero de clusters (k)')
    ax1.set_ylabel('Inercia')
    ax1.grid(True)

    # Gr√°fico del Silhouette score
    ax2.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
    ax2.set_title('Silhouette Score')
    ax2.set_xlabel('N√∫mero de clusters (k)')
    ax2.set_ylabel('Silhouette Score')
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

    return inertias, silhouette_scores

def apply_clustering(X, n_clusters=3):
    """
    Aplica K-means clustering de dos formas:
    1. K-means est√°ndar sobre los datos originales
    2. K-means sobre datos reducidos con PCA

    Args:
        X: matriz de datos estandarizados
        n_clusters: n√∫mero de clusters a crear

    Returns:
        tuple: (kmeans_labels, kmeans_pca_labels, X_pca, pca_explained_variance_ratio)
    """
    # 1. K-means est√°ndar
    print("Aplicando K-means est√°ndar...")
    kmeans = KMeans(n_clusters=n_clusters, random_state=RANDOM_STATE)
    kmeans_labels = kmeans.fit_predict(X)

    # 2. PCA + K-means
    print("Aplicando PCA + K-means...")

    # Determinamos el n√∫mero m√°ximo de componentes posible
    max_components = min(X.shape[0], X.shape[1])

    # Aplicamos PCA inicialmente con todos los componentes
    pca_full = PCA(n_components=max_components, random_state=RANDOM_STATE)
    pca_full.fit(X)

    # Calculamos el n√∫mero de componentes necesarios para 95% de varianza
    cumsum = np.cumsum(pca_full.explained_variance_ratio_)
    n_components_95 = np.argmax(cumsum >= 0.95) + 1
    n_components_95 = min(n_components_95, max_components)  # Aseguramos no exceder el m√°ximo

    # Aplicamos PCA con el n√∫mero correcto de componentes
    pca = PCA(n_components=n_components_95, random_state=RANDOM_STATE)
    X_pca_reduced = pca.fit_transform(X)
    print(f"N√∫mero de componentes seleccionados por PCA (95% varianza): {n_components_95}")

    # Aplicamos K-means sobre los datos reducidos
    kmeans_pca = KMeans(n_clusters=n_clusters, random_state=RANDOM_STATE)
    kmeans_pca_labels = kmeans_pca.fit_predict(X_pca_reduced)

    # Reducimos a 2D solo para visualizaci√≥n
    pca_2d = PCA(n_components=2, random_state=RANDOM_STATE)
    X_pca_2d = pca_2d.fit_transform(X_pca_reduced)

    return kmeans_labels, kmeans_pca_labels, X_pca_2d, pca_2d.explained_variance_ratio_

def visualize_clusters(X, kmeans_labels, kmeans_pca_labels, X_pca):
    """
    Visualiza los resultados del clustering usando t-SNE y PCA.

    Args:
        X: matriz de datos original
        kmeans_labels: etiquetas del K-means est√°ndar
        kmeans_pca_labels: etiquetas del K-means con PCA
        X_pca: datos transformados por PCA (2D)
    """
    # Preparar visualizaci√≥n con t-SNE
    print("Preparando visualizaci√≥n t-SNE...")
    tsne = TSNE(n_components=2, random_state=RANDOM_STATE)
    X_tsne = tsne.fit_transform(X)

    # Crear subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # 1. Visualizaci√≥n t-SNE del K-means est√°ndar
    scatter1 = ax1.scatter(X_tsne[:, 0], X_tsne[:, 1], c=kmeans_labels, cmap='viridis')
    ax1.set_title('Clusters (K-means) visualizados con t-SNE')
    ax1.set_xlabel('t-SNE 1')
    ax1.set_ylabel('t-SNE 2')
    ax1.grid(True)
    plt.colorbar(scatter1, ax=ax1, label='Cluster')

    # 2. Visualizaci√≥n directa del PCA + K-means
    scatter2 = ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_pca_labels, cmap='viridis')
    ax2.set_title('Clusters (PCA + K-means)')
    ax2.set_xlabel('Componente Principal 1')
    ax2.set_ylabel('Componente Principal 2')
    ax2.grid(True)
    plt.colorbar(scatter2, ax=ax2, label='Cluster')

    plt.tight_layout()
    plt.show()

def main():
    """
    Funci√≥n principal que ejecuta el an√°lisis completo de clustering.
    """
    # 1. Cargar y preprocesar datos
    print("\n=== Inicio del An√°lisis de Clustering ===")
    X_scaled, feature_names = load_and_preprocess_data()

    # 2. Encontrar n√∫mero √≥ptimo de clusters
    print("\n=== An√°lisis del N√∫mero √ìptimo de Clusters ===")
    inertias, silhouette_scores = find_optimal_clusters(X_scaled)

    # 3. Aplicar clustering
    print("\n=== Aplicaci√≥n de K-means ===")
    optimal_clusters = 3  # Elegido basado en el an√°lisis anterior
    kmeans_labels, kmeans_pca_labels, X_pca, explained_variance = apply_clustering(
        X_scaled, optimal_clusters)

    # 4. Visualizar resultados
    print("\n=== Visualizaci√≥n de Resultados ===")
    visualize_clusters(X_scaled, kmeans_labels, kmeans_pca_labels, X_pca)

    # 5. Imprimir m√©tricas finales
    print("\n=== M√©tricas de Rendimiento ===")
    print(f"Silhouette Score (K-means): {silhouette_score(X_scaled, kmeans_labels):.3f}")
    print(f"Silhouette Score (PCA + K-means): {silhouette_score(X_pca, kmeans_pca_labels):.3f}")
    print(f"Varianza explicada por PCA: {sum(explained_variance)*100:.2f}%")

# Punto de entrada del script
if __name__ == "__main__":
    main()
```
